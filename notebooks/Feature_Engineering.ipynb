### Feature Engineering & Train/ Test Split

**Purpose:** Select the most relevant feature, prepare the data for modeling, and split into training and test sets.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import os
import warnings

warnings.filterwarnings("ignore")

PROCESSED_DIR = "../data/processed"

# Loading cleaned UCI dataset
df = pd.read_csv(os.path.join(PROCESSED_DIR, "uci_credit_default_cleaned.csv"))

# Selecting necessary features for modeling
# pay_0 through pay_6 = payment status last 6 months (strongest predictors)
# limit_bal = credit limit (strongest from heatmap)
# bill_amt1 only (dropping amt2/amt3 - highly collinear per Section 2 findings)
# pay_amt1 only (same reason as the one before)
# age, education, marriage = demographic features

features = [
    "limit_bal",
    "age",
    "education",
    "marriage",
    "pay_0", "pay_2", "pay_3",
    "bill_amt1",
    "pay_amt1"
]

target = "delinquent"

X = df[features]
y = df[target]

print(f"Features selected : {X.shape[1]}")
print(f"Total records     : {X.shape[0]:,}")
print(f"Delinquency rate  : {y.mean()*100:.1f}%")
print(f"\nFeature columns   : {list(X.columns)}")
```

```python
# Scale features so all values are on the same range

scaler = StandardScaler()   # transforming each feature to have mean=0 and Std. Dev = 1
X_scaled = scaler.fit_transform(X)   # fit -> calculates the mean and std dev of each col,
                                        # transform -> applies the scaling

# Convert back to DataFrame so column names are preserved
X_scaled = pd.DataFrame(X_scaled, columns=features) #return the plain Numpy array to wrap back into the DF

# Split into 80% training and 20% test
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y,
    test_size=0.20,
    random_state=42
)

print(f"Training set   : {X_train.shape[0]:,} records")
print(f"Test set       : {X_test.shape[0]:,} records")
print(f"\nTrain delinquency rate : {y_train.mean()*100:.1f}%")
print(f"Test delinquency rate  : {y_test.mean()*100:.1f}%")
```

```python
# Save all four splits to processed folder for use in Section 4
X_train.to_csv(os.path.join(PROCESSED_DIR, "X_train.csv"), index=False)
X_test.to_csv(os.path.join(PROCESSED_DIR, "X_test.csv"), index=False)
y_train.to_csv(os.path.join(PROCESSED_DIR, "y_train.csv"), index=False)
y_test.to_csv(os.path.join(PROCESSED_DIR, "y_test.csv"), index=False)

print("Train/test sets saved to data/processed/")
```
